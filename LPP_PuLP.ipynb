# Linear Programming Problem with 2 variables using Pulp library


### Minimization 

#### Importing Libraries

```python
import pulp as p
import numpy as np
import matplotlib.pyplot as plt
```

#### Defining object of the LPP

```python
Lp_prob = p.LpProblem('Problem1', p.LpMinimize)
```

#### Decision Variables

```python
x1 = p.LpVariable("x1", lowBound = 0) # Create a variable x1 >= 0
x2 = p.LpVariable("x2", lowBound = 0) # Create a variable x2 >= 0
```

#### Objective Function

```python
Lp_prob += 60*x1 + 40*x2       # Minimize Z = 60x1 + 40x2
```

#### Constraints

```python
Lp_prob += 4*x1 + 4*x2 >= 10   # 4x1 + 4x2 >= 10
Lp_prob += 2*x1 + x2 >= 4      # 2x1 + x2 >= 4
Lp_prob += 6*x1 + 2*x2 <= 12   # 6x1 + 2x2 <= 12
print(Lp_prob)
```

#### Solving the LPP

```python
status = Lp_prob.solve()
print(f"status : {p.LpStatus[status]}")
m = p.value(x1)
n = p.value(x2)
print("x1 =", m) 
print("x2 =", n)
print("The optimal solution is : ", p.value(Lp_prob.objective))
```

#### Plotting the Graphical solution

```python
# Minimize Z = 60x1 + 40x2
x1 = np.linspace(0, 20, 2000)          

# Make plot
x2_1 = 2.5 - x1      # 4x1 + 4x2 >= 10
x2_2 = 4 - 2*x1      # 2x1 + x2 >= 4
x2_3 = 6 - 3*x1      # 6x1 + 2x2 <= 12
    
plt.plot(x1, x2_1, label='4x1 + 4x2 >= 10')
plt.plot(x1, x2_2, label='2x1 + x2 >= 4')
plt.plot(x1, x2_3, label='6x1 + 2x2 <= 12')

plt.xlim((0, 3))
plt.ylim((0, 6))
plt.xlabel('x1', labelpad = 10)
plt.ylabel('x2', rotation = 0, labelpad = 15)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

# Fill feasible region
upper = x2_3
lower = np.maximum(x2_1, x2_2)
plt.fill_between(x1, lower, upper, where = lower < upper, color='grey', alpha=0.5)
plt.text(0.1, 4, 'Feasible \n Region', size = '11')
plt.annotate(f'Optimal \n solution\n({m}, {n})', xy = (m, n))
plt.grid(True)
plt.show()
```

# Linear Programming Problem with 2 variables without using Pulp library


### Minimization

#### Importing Libraries

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
```

#### Constraints

```python
# Minimize Z = x1 + 2x2
eq1 = [2,1, 2]         # 2x1 + x2 >= 2
eq2 = [3,4, 12]        # 3x1 + 4x2 <= 12
eq3 = [0,np.inf, 0]    # x2 >= 0
eq4 = [np.inf, 0,0]    # x1 >= 0
arr = np.array([eq1,eq2,eq3,eq4])
```

#### Function to find intersection points

```python
def intersection_points(eq1, eq2):
    sol1, sol2 = eq1[-1], eq2[-1]
    A = np.array([eq1[:-1],eq2[:-1]])
    b = np.array([sol1,sol2])
    x1,x2 = np.dot(np.linalg.inv(A),b)
    return (x1,x2)
```

#### Finding the intersection points and solution

```python
labels = []
opt = np.inf
for i in range(4):
    for j in range(i+1,4):
        try:
            x1, x2 = intersection_points(arr[i],arr[j])
            if (2*x1 + x2 >= 2) & (3*x1 + 4*x2 <= 12):
                print("Intersection point for eq",(i+1), "and eq",(j+1),"is: ",(x1,x2),
                      "And the Value is: ",x1 + 2*x2)
                labels.append((x1,x2))
                if x1 + 2*x2 < opt:
                    opt = x1 + 2*x2
                    m,n = x1,x2
        except:
            print("eq",(i+1), "and eq",(j+1),"has No Intersection Point")
            
print("\nThe Optimal Value is ",opt, "at point ",(m,n))
```

#### Plotting the Graphical solution

# Write a code to implement Particle Swarm Optimization for sphere function without constraints.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the objective function (Sphere function)
def sphere(position):
    return np.sum(np.square(position))

# Define limits for position and velocity
def limitX(X, xMin, xMax):
    for i in range(len(X)):
        if X[i] > xMax:
            X[i] = xMax
        if X[i] < xMin:
            X[i] = xMin
    return X

def limitv(V, vMin, vMax):
    for i in range(len(V)):
        if V[i] > vMax:
            V[i] = vMax
        if V[i] < vMin:
            V[i] = vMin
    return V

# PSO Implementation
def optimization():
    class Particle:
        def __init__(self, ps, d, xMin, xMax, vMin, vMax, MaxIt, c1, c2, w):
            self.ps = ps
            self.d = d
            self.xMin = xMin
            self.xMax = xMax
            self.vMin = vMin
            self.vMax = vMax
            self.MaxIt = MaxIt
            self.c1 = c1
            self.c2 = c2
            self.w = w
            
            # Initialize particle positions and velocities
            self.position = np.random.uniform(self.xMin, self.xMax, [self.ps, self.d])
            self.velocity = np.random.uniform(self.vMin, self.vMax, [self.ps, self.d])
            self.cost = np.apply_along_axis(sphere, 1, self.position)
            
            # Initialize personal best and global best
            self.pbest = np.copy(self.position)
            self.pbest_cost = np.copy(self.cost)
            self.index = np.argmin(self.pbest_cost)
            self.gbest = self.pbest[self.index]
            self.gbest_cost = self.pbest_cost[self.index]
            
            # Store best cost per iteration
            self.BestCost = np.zeros(self.MaxIt)

        def Evaluate(self):
            for it in range(self.MaxIt):
                for i in range(self.ps):
                    # Update velocity
                    self.velocity[i] = (
                        self.w[it] * self.velocity[i]
                        + self.c1 * np.random.rand(self.d) * (self.pbest[i] - self.position[i])
                        + self.c2 * np.random.rand(self.d) * (self.gbest - self.position[i])
                    )
                    # Apply velocity limits
                    self.velocity[i] = limitv(self.velocity[i], self.vMin, self.vMax)
                    
                    # Update position
                    self.position[i] = self.position[i] + self.velocity[i]
                    self.position[i] = limitX(self.position[i], self.xMin, self.xMax)
                    
                    # Evaluate cost
                    self.cost[i] = sphere(self.position[i])
                    
                    # Update personal best
                    if self.cost[i] < self.pbest_cost[i]:
                        self.pbest[i] = self.position[i]
                        self.pbest_cost[i] = self.cost[i]
                        
                        # Update global best
                        if self.pbest_cost[i] < self.gbest_cost:
                            self.gbest = self.pbest[i]
                            self.gbest_cost = self.pbest_cost[i]
                
                # Store best cost for this iteration
                self.BestCost[it] = self.gbest_cost

        def plot(self):
            # Plot the convergence curve
            plt.semilogy(self.BestCost)
            plt.ylim([10e-120, 10e20])
            plt.xlim([0, self.MaxIt])
            plt.ylabel("Best Function Value")
            plt.xlabel("Number of Iterations")
            plt.title("Particle Swarm Optimization of Sphere Function")
            print("Best fitness value : ",self.gbest_cost)
            plt.grid(True)
            
            # Plotting the 3D surface plot of sphere function
            
            # Meshgrid creation
            x = np.arange(-10, 10, 0.05)
            y = np.arange(-10, 10, 0.05)
            x_new, y_new = np.meshgrid(x, y)
            
            # Evaluate the objective function on the meshgrid
            z = np.zeros_like(x_new)
            for i in range(x_new.shape[0]):
                for j in range(x_new.shape[1]):
                    current_x = np.array([x_new[i, j], y_new[i, j]])
                    z[i, j] = sphere(current_x)
            
            fig = plt.figure(figsize = (10,7))
            ax = fig.add_subplot(111, projection='3d')
            ax.plot_surface(x_new, y_new, z, cmap='viridis', alpha=0.8)
            ax.set_xlabel('x1')
            ax.set_ylabel('x2')
            ax.set_zlabel('Objective Function Value')
            ax.set_title("Surface plot of sphere function")
            plt.show()

    # PSO parameters
    ps = 10  # Population size
    d = 10    # Dimension of the problem
    xMin, xMax = -100, 100
    vMin, vMax = -0.2 * (xMax - xMin), 0.2 * (xMax - xMin)
    MaxIt = 3000
    c1, c2 = 2.0, 2.0
    w = 0.9 - ((0.9 - 0.4) / MaxIt) * np.linspace(0, MaxIt, MaxIt)  # Inertia weight (linearly decreasing)

    # Run PSO
    pso = Particle(ps, d, xMin, xMax, vMin, vMax, MaxIt, c1, c2, w)
    pso.Evaluate()
    pso.plot()

# Execute the optimization
optimization()
```

# Write a code to implement Particle Swarm Optimization for sphere function with constraint.

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    # Define the objective function with constraints
    const1 = x[1] <= 3.2 or x[1] >= 6.4
    const2 = (x[0]**2 + x[1]**2) >= 1
    const3 = x[0] != x[1]
    
    if const1 and const2 and const3:
        o = np.sum(x**2)
    else:
        o = np.sum(x**2) + 200  # Penalty for not satisfying constraints

    return o
    
def initialize_particles(num_particles, dim, bounds):
    particles = np.random.uniform(bounds[0], bounds[1], size=(num_particles, dim))
    velocities = np.random.rand(num_particles, dim)
    best_positions = np.copy(particles)
    best_costs = np.full(num_particles, float('inf'))
    return particles, velocities, best_positions, best_costs

def update_velocity(velocity, inertia, c1, c2, particle_position, best_position, global_best_position):
    return (
        inertia * velocity 
        + c1 * np.random.rand(len(particle_position)) * (best_position - particle_position) 
        + c2 * np.random.rand(len(particle_position)) * (global_best_position - particle_position)
    )

def update_position(position, velocity):
    return position + velocity

def pso(num_particles, dim, bounds, max_iter, inertia, c1, c2):
    particles, velocities, best_positions, best_costs = initialize_particles(num_particles, dim, bounds)
    global_best_position = None
    global_best_cost = float('inf')
    best_costs_list = []

    for iteration in range(max_iter):
        for i in range(num_particles):
            cost = objective_function(particles[i])
            if cost < best_costs[i]:
                best_costs[i] = cost
                best_positions[i] = np.copy(particles[i])
            if cost < global_best_cost:
                global_best_cost = cost
                global_best_position = np.copy(particles[i])

        for i in range(num_particles):
            velocities[i] = update_velocity(
                velocities[i], inertia, c1, c2, particles[i], best_positions[i], global_best_position
            )
            particles[i] = update_position(particles[i], velocities[i])
        
        best_costs_list.append(max(global_best_cost, 1e-10))  # Ensure cost is positive for plotting
        #print(f'Iteration {iteration + 1}: Best Cost = {global_best_cost}')

    return global_best_position, global_best_cost, best_costs_list

def plot(best_costs_list, global_best_cost, max_iter):
    # Plot the convergence curve    
    plt.semilogy(best_costs_list)  # Logarithmic plot for convergence
    plt.ylim([1e-1, 2000])   # Adjust y-axis limits for better visualization
    plt.xlim([0, max_iter])   # Set x-axis limit based on max_iter
    plt.ylabel("Best Function Value")
    plt.xlabel("Number of Iterations")
    plt.title("Convergence of Particle Swarm Optimization")
    print("Best fitness value:", global_best_cost)
    plt.grid(True)
    
    # Meshgrid creation for visualization
    x = np.arange(-10, 10, 0.05)
    y = np.arange(-10, 10, 0.05)
    x_new, y_new = np.meshgrid(x, y)
    
    z = np.zeros_like(x_new)
    for i in range(x_new.shape[0]):
        for j in range(x_new.shape[1]):
            current_x = np.array([x_new[i, j], y_new[i, j]])
            z[i, j] = objective_function(current_x)  # Objective function values for surface plot

    # 3D Surface plot
    fig = plt.figure(figsize = (10,7))
    ax3d = fig.add_subplot(111, projection='3d')
    ax3d.plot_surface(x_new, y_new, z, cmap='viridis', alpha=0.8)
    ax3d.set_xlabel('x1')
    ax3d.set_ylabel('x2')
    ax3d.set_zlabel('Objective Function Value')
    ax3d.set_title("Surface Plot of Objective Function")
    plt.show()

# Example usage
num_particles = 10
dim = 2  # Assuming a 2D optimization problem for the visualization
bounds = (-100, 100)
max_iter = 1000
inertia = 0.5
c1 = 2
c2 = 2

global_best_position, global_best_cost, best_costs_list = pso(
    num_particles, dim, bounds, max_iter, inertia, c1, c2
)
plot(best_costs_list, global_best_cost, max_iter)
```

# Write a code to implement Binary Particle Swarm Optimization for sphere function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid Activation Function
def sigmoid(x):
    x = np.array(x)  # Ensure x is a numpy array
    return 1 / (1 + np.exp(-x))

# Objective Function (Sphere function as an example)
def objective_function(x):
    return np.sum(np.square(x))

# Initialize the PSO parameters
nVar = 20  # Number of dimensions (variables)
bounds = [0, 1]  # Binary bounds for BPSO
noP = 1000  # Number of particles
maxIter = 500  # Maximum number of iterations
wMax = 0.9  # Maximum inertia weight
wMin = 0.2  # Minimum inertia weight
c1 = 2  # Cognitive coefficient
c2 = 2  # Social coefficient
vMax = 0.2  # Maximum velocity
vMin = -vMax  # Minimum velocity

# Initialize the particles
particles = []
for _ in range(noP):
    particle = {
        "X": np.random.randint(0, 2, nVar),  # Random binary positions (0 or 1)
        "V": np.zeros(nVar),  # Velocity initialized to zero
        "PBEST_X": np.zeros(nVar),  # Personal best position
        "PBEST_O": float('inf')  # Personal best objective value
    }
    particles.append(particle)

# Initialize global best
GBEST_X = np.zeros(nVar)
GBEST_O = float('inf')

# Store convergence curve for plotting
cgCurve = []

# Main optimization loop
for t in range(maxIter):
    w = wMax - t * ((wMax - wMin) / maxIter)  # Inertia weight adjustment

    for particle in particles:
        currentX = particle["X"]
        currentO = objective_function(currentX)

        # Update PBEST
        if currentO < particle["PBEST_O"]:
            particle["PBEST_X"] = currentX.copy()
            particle["PBEST_O"] = currentO

        # Update GBEST
        if currentO < GBEST_O:
            GBEST_X = currentX.copy()
            GBEST_O = currentO

        # Update velocity and position
        r1 = np.random.rand(nVar)
        r2 = np.random.rand(nVar)
        cognitive = c1 * r1 * (particle["PBEST_X"] - particle["X"])
        social = c2 * r2 * (GBEST_X - particle["X"])
        particle["V"] = w * particle["V"] + cognitive + social

        particle["V"] = np.clip(particle["V"], vMin, vMax)

        s = sigmoid(particle["V"])
        particle["X"] = np.where(np.random.rand(nVar) < s, 1, 0)

    cgCurve.append(GBEST_O)
    #print(f"Iteration {t + 1}: Best Objective = {GBEST_O}")

# Plot the convergence curve
plt.figure(figsize=(10, 6))
plt.plot(cgCurve, label='Best Objective Value')
plt.xlabel('Iteration')
plt.ylabel('Objective Value')
plt.title('Convergence Curve for BPSO')
plt.legend()
plt.grid(True)
plt.show()

print("\nOptimal Solution Found:")
print(f"Position: {GBEST_X}")
print(f"Best Objective Value: {GBEST_O}")
```

# Write a code to implement Discrete Particle Swarm Optimization for sphere function.

```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    return np.sum(np.square(x))

# Initialize the DPSO parameters
nVar = 5  # Number of dimensions (variables)
Discrete_set = np.array([10, 20, 30, 40])  # Discrete set of values
noP = 30  # Number of particles
maxIter = 300  # Maximum number of iterations
wMax = 0.9  # Maximum inertia weight
wMin = 0.2  # Minimum inertia weight
c1 = 2  # Cognitive coefficient
c2 = 2  # Social coefficient
vMax = 1  # Maximum velocity index change
vMin = -vMax  # Minimum velocity index change

# Initialize the particles
particles = []
for _ in range(noP):
    particle = {
        "X": np.random.choice(Discrete_set, nVar),  # Random initial positions from Discrete_set
        "V": np.zeros(nVar),  # Velocity initialized to zero
        "PBEST_X": np.zeros(nVar),  # Personal best position
        "PBEST_O": float('inf')  # Personal best objective value
    }
    particles.append(particle)

# Initialize global best
GBEST_X = np.zeros(nVar)
GBEST_O = float('inf')

# Store convergence curve for plotting
cgCurve = []

# Main optimization loop
for t in range(maxIter):
    w = wMax - t * ((wMax - wMin) / maxIter)  # Inertia weight adjustment

    for particle in particles:
        currentX = particle["X"]
        currentO = objective_function(currentX)

        # Update PBEST
        if currentO < particle["PBEST_O"]:
            particle["PBEST_X"] = currentX.copy()
            particle["PBEST_O"] = currentO
        if currentO < GBEST_O:
            GBEST_X = currentX.copy()
            GBEST_O = currentO

        # Update velocity and position
        r1 = np.random.rand(nVar)
        r2 = np.random.rand(nVar)
        cognitive = c1 * r1 * (particle["PBEST_X"] - particle["X"])
        social = c2 * r2 * (GBEST_X - particle["X"])
        particle["V"] = w * particle["V"] + cognitive + social
        particle["V"] = np.clip(particle["V"], vMin, vMax)

        for i in range(nVar):
            idx = np.where(Discrete_set == particle["X"][i])[0][0] 
            new_idx = int(idx + particle["V"][i])  # Update index by velocity
            new_idx = np.clip(new_idx, 0, len(Discrete_set) - 1)  
            particle["X"][i] = Discrete_set[new_idx]    # Update position to the new discrete value

    cgCurve.append(GBEST_O)
    #print(f"Iteration {t + 1}: Best Objective = {GBEST_O}")

# Plot the convergence curve
plt.figure(figsize=(10, 6))
plt.plot(cgCurve, label='Best Objective Value')
plt.xlabel('Iteration')
plt.ylabel('Objective Value')
plt.title('Convergence Curve for DPSO with Discrete Set')
plt.legend()
plt.grid(True)
plt.show()

print("\nOptimal Solution Found:")
print(f"Position: {GBEST_X}")
print(f"Best Objective Value: {GBEST_O}")
```
